```python
import numpy as np
import matplotlib.pyplot as plt

# t goes from 0 to 2*pi
t = np.linspace(0, 2 * np.pi, 1000)

# Parametric equations for a heart
x = 16 * np.sin(t)**3
y = 12 * np.cos(t) - 5 * np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

plt.figure(figsize=(7,7))
plt.plot(x, y, color='red')
plt.fill(x, y, color='red')  # correct fill
plt.axis('off')
plt.title("I love you Letkaaaaaaaaaaa <3", fontsize=16)
plt.show()
```
